```python
# Selecting possible columns for analysis

columns_selected_names = ['Price', 'Postal Code', 'Build Year', 'Facades', 'Habitable Surface', 'Land Surface', 'Type', 'Subtype', 'Bedroom Count', 'Bathroom Count', 'Toilet Count', 'Room Count', 'Kitchen Surface', 'Furnished', 'Living Surface']
```

```python
# Creating a new dataframe with the selected columns
immo_extracted = immo_df[columns_selected_names]
immo_extracted.head()
```

```python
correlation_matrix = immo_extracted[
    ['Price', 'Postal Code', 'Build Year', 'Facades', 'Habitable Surface', 'Land Surface', 'Bedroom Count',
     'Bathroom Count', 'Toilet Count', 'Room Count', 'Kitchen Surface', 'Living Surface']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Inspecting for outliers in Price
plt.figure(figsize=(10, 4)) 
plt.boxplot(immo_extracted['Price'])   
plt.title('Price')           
plt.ylabel('Values')       
plt.tight_layout()         
plt.show()                  
```

```python
# Calculating the interquartile range (IQR)
Q1 = immo_extracted['Price'].quantile(0.25)
Q3 = immo_extracted['Price'].quantile(0.75)
IQR = Q3 - Q1

# the limits to detect outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Identify outliers
outliers = immo_extracted[(immo_extracted['Price'] < lower_limit) | (immo_extracted['Price'] > upper_limit )]

# Calculates the number of values inside the IQR and the number of outliers.
values_inside_IQR = immo_extracted[(immo_extracted['Price'] >= lower_limit) & (immo_extracted['Price'] <= upper_limit)].shape[0]
quantity_outliers = outliers.shape[0]

print(f"Price mean: {immo_extracted['Price'].mean()}")
print(f"Outliers Price mean: {outliers['Price'].mean()}")

print(f"Outliers Count: \n{outliers.count()}")
```

```python
columns_selected_names = ['Price', 'Subtype']

for property_type in immo_extracted['Subtype'].unique():
    plt.figure(figsize=(8, 6))  
    data_subset = immo_extracted[immo_extracted['Subtype'] == property_type]
    plt.boxplot(data_subset['Price']) 
    plt.title(property_type) 
    plt.ylabel('Price') 
    plt.xticks([1], [property_type])
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
sns.boxplot(data=immo_extracted, x='Postal Code', y='Price')
```

```python
# Sorting zip codes by median price
sorted_postal_codes = immo_extracted.groupby('Postal Code')['Price'].median().sort_values(ascending=False).index

# Limit the number of displayed zip codes
n_top_postal_codes = 30
top_postal_codes = sorted_postal_codes[:n_top_postal_codes]

# Filter the DataFrame to include only the selected zip codes
immo_filtered = immo_extracted[immo_extracted['Postal Code'].isin(top_postal_codes)]

# Configure the size of the figure
plt.figure(figsize=(12, 6))

# Create the box plot
sns.boxplot(data=immo_filtered, x='Postal Code', y='Price')

# Rotate x-axis ticks to improve readability
plt.xticks(rotation=45)

# Add title and axis labels
plt.title('Price by Zip Code Boxplot')
plt.xlabel('Zip Code')
plt.ylabel('Price')

# Show the graph
plt.tight_layout()
plt.show()
```
