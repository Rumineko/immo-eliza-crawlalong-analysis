```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
immo_df = pd.read_csv('../data/cleaned/appended_data.csv')
```

```python
map = {
    'TO_RESTORE': 1, 
    'TO_RENOVATE': 2, 
    'TO_BE_DONE_UP': 3, 
    'GOOD': 4,
    'JUST_RENOVATED': 5,
    'AS_NEW': 6
}
immo_df['State of Building'] = immo_df['State of Building'].apply(lambda x: map.get(x, np.NAN))
```

```python
kitchen_types={
    "NOT_INSTALLED": 1,
    "USA_UNINSTALLED": 1,
    "SEMI_EQUIPPED": 2,
    "USA_SEMI_EQUIPPED": 2,
    "INSTALLED": 3,
    "USA_INSTALLED": 3,
    "HYPER_EQUIPPED": 4,
    "USA_HYPER_EQUIPPED": 4,
}
immo_df['Kitchen Type'] = immo_df['Kitchen Type'].apply(lambda x: kitchen_types.get(x, np.NAN))
```

```python
energy_ratings = {
    "G": 0,
    "F": 1,
    "E": 2,
    "D": 3,
    "C": 4,
    "B": 5,
    "A": 6,
    "A+": 7,
    "A++": 8
}
immo_df['EPC'] = immo_df['EPC'].apply(lambda x: energy_ratings.get(x, np.NAN))
```

```python
# in all the columns replace the NaN values with the mean of the column
```

```python
include_in_matrix = ['Price', 'Habitable Surface', 'Build Year', 'State of Building', 'EPC', "Kitchen Type", 'State of Building', ]
['Land Surface', 'Swimming Pool']
['Terrace Surface']
corr_matrix = immo_df[include_in_matrix].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
